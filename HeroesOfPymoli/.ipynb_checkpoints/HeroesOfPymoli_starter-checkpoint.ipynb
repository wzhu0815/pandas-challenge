## Gender Demographics

* Percentage and Count of Male Players

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0,9,14,19,24,29,34,39,100]
group_names = ["<10","10-14","15-19",'20-24','25-29','30-34','35-39','40+']
purchase_data['Age Ranges'] = pd.cut(purchase_data['Age'],bins, labels = group_names,include_lowest = True)
tc_gp = purchase_data.groupby('Age Ranges')
price_count_ds = tc_gp['Price'].count()
price_ave_ds = tc_gp['Price'].mean().map("${:.2f}".format)
price_sum_ds = tc_gp['Price'].sum()
pa_df = pd.concat([price_count_ds,price_ave_ds,price_sum_ds],axis=1)
pa_df.columns= ["Purchase Count","Average Purchase Price", "Total Purchase Value"]
pa_df['Avg Total Purchase per Person'] = (pa_df['Total Purchase Value']/age_df['Total Count']).map("${:.2f}".format)
pa_df['Total Purchase Value'] = pa_df['Total Purchase Value'].map("${:.2f}".format)
pa_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_ds = purchase_data.groupby("SN")['Price'].sum().sort_values(ascending = False)
top5_ls = top_ds.index[0:5] #https://stackoverflow.com/questions/18327624/find-elements-index-in-pandas-series
filt =(purchase_data['SN'].isin(top5_ls))
top5_ds = purchase_data.loc[filt, ['SN','Price']]
top5_gp = top5_ds.groupby('SN')
pc_ds = top5_gp['Price'].count()
a_p_ds = top5_gp['Price'].mean().map("${:.2f}".format)
t_p_ds = top5_gp['Price'].sum().map("${:.2f}".format)
top5_df = pd.concat([pc_ds,a_p_ds,t_p_ds], axis=1)
top5_df.columns = ["Purchase Count","Average Purchase Price", "Total Purchase Value"]
top5_df = top5_df.sort_values('Total Purchase Value',ascending = False)
top5_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_data.columns
item_gp = purchase_data[['Item ID','Item Name','Price']].groupby(['Item ID','Item Name'])
pc_ds = item_gp['Price'].count()
it_ds = item_gp['Price'].mean().map("${:.2f}".format)
tpv_ds = item_gp['Price'].sum().map("${:.2f}".format)
item_df = pd.concat([pc_ds,it_ds,tpv_ds],axis=1)
item_df.columns = ["Purchase Count","Item Price", "Total Purchase Value"]
item_df = item_df.sort_values('Purchase Count',ascending = False)
item_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
item_gp = purchase_data[['Item ID','Item Name','Price']].groupby(['Item ID','Item Name'])
pc_ds2 = item_gp['Price'].count()
it_ds2 = item_gp['Price'].mean().map("${:.2f}".format)
tpv_ds2 = item_gp['Price'].sum()
item_df2 = pd.concat([pc_ds2,it_ds2,tpv_ds2],axis=1)
item_df2.columns = ["Purchase Count","Item Price", "Total Purchase Value"]
item_df2 = item_df2.sort_values('Total Purchase Value',ascending = False)
item_df2['Total Purchase Value'] = item_df2['Total Purchase Value'].map("${:.2f}".format)
item_df2.head()
```
