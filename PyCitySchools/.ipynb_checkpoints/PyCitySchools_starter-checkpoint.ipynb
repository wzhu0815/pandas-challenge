```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
t_schools = len(school_data_complete['School ID'].unique())
t_students = "{:,}".format(school_data_complete['Student ID'].count())
t_budget = "${:,.2f}".format(school_data['budget'].sum())
a_m_score = "{:.6f}".format(school_data_complete['math_score'].mean())
a_r_score = "{:.5f}".format(school_data_complete['reading_score'].mean())
# The passing score in USA is 70?
m_passing = "{:.6f}".format((school_data_complete['School ID'][school_data_complete['math_score']>=70].count())/(school_data_complete['School ID'].count())*100)
r_passing = "{:.6f}".format((school_data_complete['School ID'][school_data_complete['reading_score']>=70].count())/(school_data_complete['School ID'].count())*100)
o_passing = "{:.6f}".format((school_data_complete['School ID'][(school_data_complete['reading_score']>=70)&(school_data_complete['math_score']>=70)].count())/(school_data_complete['School ID'].count())*100)
```

```python
ds_df = pd.DataFrame([{"Total Schools":t_schools,
                      "Total Students":t_students,
                      "Total Budget":t_budget,
                      "Average Math Score":a_m_score,
                     "Average Reading Score":a_r_score,
                     "% Passing Math":m_passing,
                     "% Passing Reading":r_passing,
                     "% Overall Passing":o_passing}])
ds_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
ss_df = school_data.set_index("school_name")
school_data_complete2 = school_data_complete.copy()
school_data_complete2['Pass Math'] = (school_data_complete2['math_score']>=70)
school_data_complete2['Pass Reading'] = (school_data_complete2['reading_score']>=70)
school_data_complete2['Pass overall'] = ((school_data_complete2['reading_score']>=70)&(school_data_complete2['math_score']>=70))
```

```python
ss_gp = school_data_complete2.groupby('school_name')
ss_df2 = pd.DataFrame()
ss_df2['School Type'] = ss_df['type']
ss_df2['Total Students'] = ss_gp['Student ID'].count()
ss_df2['Total School Budget'] = ss_df['budget'].map("${:,.2f}".format)
ss_df2['Per Student Budget'] = (ss_df['budget']/ss_df2['Total Students']).map("${:,.2f}".format)

ss_df2['Average Math Score'] = ss_gp['math_score'].mean()
ss_df2['Average Reading Score'] = ss_gp['reading_score'].mean()
ss_df2['% Passing Math'] = ss_gp['Pass Math'].sum()/ss_gp['Student ID'].count()*100
ss_df2['% Passing Reading'] = ss_gp['Pass Reading'].sum()/ss_gp['Student ID'].count()*100
ss_df2['% Overall Passing'] = ss_gp['Pass overall'].sum()/ss_gp['Student ID'].count()*100

ss_df2 = ss_df2.sort_index()
ss_df2
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
tps_df= ss_df2.sort_values('% Overall Passing',ascending = False).head()
tps_df
```

## Bottom Performing Schools (By % Overall Passing)

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
def create_ds(score,grade):
    return school_data_complete[['school_name',score]].loc[school_data_complete['grade']==grade].groupby('school_name')[score].mean()
m9_ds = create_ds('math_score','9th')
m10_ds = create_ds('math_score','10th')
m11_ds = create_ds('math_score','11th')
m12_ds = create_ds('math_score','12th')
ms_df = pd.concat([m9_ds,m10_ds,m11_ds,m12_ds], axis=1)
ms_df.columns = ["9th","10th", "11th", "12th"]
ms_df
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
def create_ds(score,grade):
    return school_data_complete[['school_name',score]].loc[school_data_complete['grade']==grade].groupby('school_name')[score].mean()
r9_ds = create_ds('reading_score','9th')
r10_ds = create_ds('reading_score','10th')
r11_ds = create_ds('reading_score','11th')
r12_ds = create_ds('reading_score','12th')
rs_df = pd.concat([r9_ds,r10_ds,r11_ds,r12_ds], axis=1)
rs_df.columns = ["9th","10th", "11th", "12th"]
rs_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
ss_gp = school_data_complete2.groupby('school_name')
ss_df3 = pd.DataFrame()
ss_df3['Per Student Budget'] = (ss_df['budget']/ss_df2['Total Students'])
ss_df3['Average Math Score'] = ss_gp['math_score'].mean()
ss_df3['Average Reading Score'] = ss_gp['reading_score'].mean()
ss_df3['% Passing Math'] = ss_gp['Pass Math'].sum()/ss_gp['Student ID'].count()*100
ss_df3['% Passing Reading'] = ss_gp['Pass Reading'].sum()/ss_gp['Student ID'].count()*100
ss_df3['% Overall Passing'] = ss_gp['Pass overall'].sum()/ss_gp['Student ID'].count()*100
ss_df3 = ss_df3.sort_index()
ss_df3 = ss_df3[['Per Student Budget','Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','% Overall Passing']]
bins = [0,585,629,644,675]
# the example table didn't cover "584"
group_names = ["<$585","$585-629","$630-644",'$645-675']
ss_df3['Spending Ranges (Per Student)'] = pd.cut(ss_df3['Per Student Budget'],bins, labels = group_names,include_lowest = True)
ssp_gp = ss_df3.groupby('Spending Ranges (Per Student)')
#style formatting for the whole dataframe: https://towardsdatascience.com/style-pandas-dataframe-like-a-master-6b02bf6468b0
ssp_df = ssp_gp[['Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','% Overall Passing']].mean().style.format("{:,.2f}")
ssp_df
```
